```python
def create_img_array(img_dir, masks, filename):
    image_rows = 420
    image_cols = 580
    
    images = os.listdir(img_dir)
    if masks != True:
        images = [i for i in images if i.endswith('.jpg')]
    total = len(images)
    
    # create empty array
    imgs = np.ndarray((total, 1, image_rows, image_cols), dtype=np.uint8)
    
    for i, image_name in enumerate(images):
        img = cv2.imread(os.path.join(img_dir, image_name), cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (image_cols, image_rows), interpolation=cv2.INTER_CUBIC)
        img = np.array([img])
        imgs[i] = img
    
    np.save(filename, imgs)
```

```python
# create features array
create_img_array(DATA_DIR, False, 'imgs_train.npy')

# create masks array
create_img_array(MASKS_DIR, True, 'imgs_mask_train.npy')
```
